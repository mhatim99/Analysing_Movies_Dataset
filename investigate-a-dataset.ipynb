### Data Cleaning (1. Removing unnecessary columns)

```python
# After discussing the structure of the data and any problems that need to be
#   cleaned, perform those cleaning steps in the second part of this section.
# Exploring the names of the columns.
movies.columns
```

```python
# Removing unnecessary columns from the dataset.
movies.drop(['cast','homepage','director','tagline','keywords','overview','production_companies', 'id', 'imdb_id', 'popularity', 'budget_adj', 'revenue_adj', 'vote_count', 'release_year'], axis = 1, inplace=True)
movies.head()
```

### Data Cleaning (2. Modifying some columns data types)

```python
# Modifying the release_date column data type to the correct type.
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies.head()
```

### Data Cleaning (3. Removing duplicates)

```python
# Detecting if there are duplicates and their count.
movies.duplicated().sum()
```

```python
# Knowing the number of rows and columns of the dataframe before removing duplicates.
movies.shape
```

```python
# Removing duplicates from the dataframe.
movies.drop_duplicates(inplace=True)
movies.shape
```

### Data Cleaning (4. Removing missing values)

```python
# I noticed the presence of 0 values in many rows in the budget and revenue columns so I'll convert them all to null values
for column in movies.columns:
    movies[column] = movies[column].replace(0, np.nan)
```

```python
# Now checking for the missing values
movies.isna().sum()
```

```python
# And now removing all the rows with missing values
movies.dropna(inplace=True)
movies.shape
```

### Research Question 1 (What are the movies that have the highest and lowest revenues?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
highest_revenue = max(movies['revenue'])
lowest_revenue = min(movies['revenue'])
highest_rev_movie = movies['original_title'].loc[movies['revenue'] == highest_revenue].iloc[0]
lowest_rev_movie = movies['original_title'].loc[movies['revenue'] == lowest_revenue].iloc[0]
print('The movie "{}" has the highest revenue with {}$, while the movie "{}" has the lowest revenue with {}$.'.format(highest_rev_movie, highest_revenue, lowest_rev_movie, lowest_revenue))
```

### Research Question 2  (What are the movies that have the highest and lowest budgets?)

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.
highest_budget = max(movies['budget'])
lowest_budget = min(movies['budget'])
highest_budg_movie = movies['original_title'].loc[movies['budget'] == highest_budget].iloc[0]
lowest_budg_movie = movies['original_title'].loc[movies['budget'] == lowest_budget].iloc[0]
print('The movie "{}" has the highest budget with {}$, while the movie "{}" has the lowest budget with {}$.'.format(highest_budg_movie, highest_budget, lowest_budg_movie, lowest_budget))
```

### Research Question 3  (What are the highest and lowest profitable movies?)

```python
# Calculating profits 
movies['profits'] = movies['revenue'] - movies['budget']

# Determining the movies with highest and lowest profits
highest_profits = max(movies['profits'])
lowest_profits = min(movies['profits'])
highest_proft_movie = movies['original_title'].loc[movies['profits'] == highest_profits].iloc[0]
lowest_proft_movie = movies['original_title'].loc[movies['profits'] == lowest_profits].iloc[0]
print('The movie "{}" has the highest profits with {}$, while the movie "{}" has the lowest profits with {}$.'.format(highest_proft_movie, highest_profits, lowest_proft_movie, lowest_profits))
```

### Research Question 4  (What are the genres of the top 10 profitable movies?)

```python
# Sorting the movies according to the profits indicating their genres
top_genres = movies.sort_values('profits', ascending=False)['genres']
top_genres.head(10)
```

### Research Question 5  (What are the most profitable genres?)

```python
# Calculating the profits mean of all movies
profits_means = movies['profits'].mean()
profits_means
```

```python
# Creating a new dataframe containing the profitable movies with profits equals to or larger than the profits mean
profitable_movies = movies[movies['profits'] >= profits_means]

# Reindexing the new dataframe
profitable_movies.index = range(len(profitable_movies))

# Previewing the new dataframe
profitable_movies.head()
```

```python
# Categories of profitable genres
profitable_movies['genres'].value_counts()
```

```python
# I noticed the presence of mixed genres together, so I'll separate them to make the arrangement more accurate
arranged_proft_movies = profitable_movies['genres'].str.cat(sep = '|')

# Making pandas series and storing the values separately
arranged_proft_movies = pd.Series(arranged_proft_movies.split('|'))
    
# Arranging the profitable genres in descending order and assigining them to a new series
arranged_proft_genres = arranged_proft_movies.value_counts(ascending = False)

# Showing the profitable genres
arranged_proft_genres
```

**As shown above we can say that the most 5 profitable genres are Comedy, Action, Drama, Adventure then Thriller.**

```python
# We can also show this in a graph as next:

arranged_proft_genres.sort_values(ascending = True, inplace = True)

# The Plot
lt = arranged_proft_genres.plot.bar(color = 'blue', fontsize = 12)

# y axis label
lt.set_ylabel('Number of the movies', fontsize = '15')

# x axis label
lt.set_xlabel('Profitable Genres', color = 'black', fontsize = '15')

# The figure size
lt.figure.set_size_inches(15, 10)

```
